```python
import numpy as np

X_neg = np.array([[0, 0], [0, 0.2], [0.2, 0.2], [0.2, 0]])
y_neg = np.array([-1, -1, -1, -1])

X_pos = np.array([[0.2, 0.1], [0.3, 0.1]])
y_pos = np.array([1, 1])

X = np.vstack((X_neg, X_pos))
y = np.concatenate((y_neg, y_pos))

# 1 (w0), x1^2, x1*x2, x2^2 (w_ij), x1^3, x1^2*x2, x1*x2^2, x2^3 (w_ijk)
def expand_features(x):
    x1, x2 = x[0], x[1]
    return [
        1,                    # w0
        x1**2, x1*x2, x2**2,  # w_ij (3个独立项)
        x1**3, (x1**2)*x2, x1*(x2**2), x2**3 # w_ijk (4个独立项)
    ]

P = np.array([expand_features(xi) for xi in X])

# pseudo
pseudo=np.linalg.inv(P@P.T)
w=P.T@pseudo@y

# 5. 验证结果
predictions = P @ w
classified = np.sign(predictions)
errors = np.sum(classified != y)

print(f"样本预测值 g(w, x): \n{predictions.round(4)}")
print(f"分类错误数: {errors}")

```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures


X = np.array([
    [0, 0], [0, 1], [1, 0], [2, 1], [0, 2], [1, 2],
    [3, 1], [2, 3], [4, 5], [5, 3], [3, 6], [6, 8]
])
y = np.array([-2.1, -1.3, -1.5, 1.0, 0.5, 0.8, 2.1, 1.8, 3.3, 2.8, 3.8, 5.0])

X_test = np.array([[0.5, 0.3], [0.8, 0.8]])


poly = PolynomialFeatures(degree=3)
P = poly.fit_transform(X)
P_t = poly.transform(X_test)

# over-determined system
w=np.linalg.inv(P.T@P)@P.T@y
y_train_pred=P@w
y_test_pred=P_t@w

# regularization lambda=0.001
lam=0.001
w_r=np.linalg.inv(P.T@P+lam*np.eye(P.shape[1]))@P.T@y
y_ridge_train_pred_r=P@w_r
y_ridge_test_pred_r=P_t@w_r

print("weight maxtrix (无正则化):", w.round(4))
print("weight maxtrix (Ridge 正则化):", w_r.round(4))
print("训练集预测值 (无正则化):", y_train_pred.round(4))
print("测试集预测值 (无正则化):", y_test_pred.round(4))
print("训练集预测值 (Ridge 正则化):", y_ridge_train_pred_r.round(4))
print("测试集预测值 (Ridge 正则化):", y_ridge_test_pred_r.round(4))
```
