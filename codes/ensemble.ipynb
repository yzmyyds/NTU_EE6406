```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# download dataset from sklearn
dataset = load_iris()
print(dataset.DESCR)
```

```python
X=dataset.data
y=dataset.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
clf2=DecisionTreeClassifier(criterion='gini',max_depth=2, random_state=42)
clf2.fit(X_train,y_train)
print(classification_report(y_test, clf2.predict(X_test), target_names=dataset['target_names']))
```

```python
#plot the decision tree
plt.figure(figsize=(10,10))
plot_tree(clf2, feature_names=dataset['feature_names'], class_names=dataset['target_names'], filled=True)
plt.show()
```
